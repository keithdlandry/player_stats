```python
%reload_ext autoreload
%autoreload 2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
pd.options.display.max_columns = 1000
pd.options.display.max_rows = 250
```

```python
positions = ['qb', 'rb', 'wr', 'te']
projections = [pd.read_csv('../projections/2015_' + pos + '_proj.csv', index_col=0) for pos in positions]
summaries = [pd.read_csv('../projections/' + pos + '_summary_table.csv', index_col=0) for pos in positions]
```

```python
merged = [pd.merge(proj, summ, on='Name', how='outer') for proj, summ in zip(projections, summaries)]
```

```python
for df in merged:
    df.drop(['2015 Projection', 'Value Above Baseline'], axis=1, inplace=True)
```

```python
for df in merged:
    df.sort(columns='MFL', ascending=True, inplace=True)
    df['ADP Rank'] = range(1,len(df)+1)
    df.sort(columns='SVM Projection', ascending=False, inplace=True)
    df['Proj Rank'] = range(1,len(df)+1)
    df['Diff'] = df['ADP Rank'] - df['Proj Rank']
    
merged[1]
```

```python
for df, pos in zip(merged, positions):
    df.to_csv('../projections/total/' + pos + '_total.csv')
```
